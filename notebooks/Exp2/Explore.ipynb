# Load Results

```python
from alabortijcv2015.utils import pickle_load

path = '/data/'
db = 'lfpw'
aam_type = 'PartsAAM'
noise_std = [None, 0, 0.02, 0.04, 0.06]

sic, sfc = [], []
aic, afc = [], []
bic, bfc, pic = [], [], []

for n in noise_std:
    sic.append(pickle_load(path + 'PhD/Results/ijcv2015/exp2_int_' + aam_type + '_SIC_' + db + '_' + str(n)))
    sfc.append(pickle_load(path + 'PhD/Results/ijcv2015/exp2_int_' + aam_type + '_SFC_' + db + '_' + str(n)))
    
    aic.append(pickle_load(path + 'PhD/Results/ijcv2015/exp2_int_' + aam_type + '_AIC_' + db + '_' + str(n)))
    afc.append(pickle_load(path + 'PhD/Results/ijcv2015/exp2_int_' + aam_type + '_AFC_' + db + '_' + str(n)))
    
    
    bic.append(pickle_load(path + 'PhD/Results/ijcv2015/exp2_int_' + aam_type + '_BIC_' + db + '_' + str(n)))
    #bfc.append(pickle_load(path + 'PhD/Results/ijcv2015/exp2_int_' + aam_type + '_BFC_' + db + '_' + str(n)))
    #pic.append(pickle_load(path + 'PhD/Results/ijcv2015/exp2_int_' + aam_type + '_PIC_' + db + '_' + str(n)))
    
results = [sic, sfc, 
           aic, afc, bic] 
           #bic, bfc, pic]
```

# Graph 1

```python
all_errors = []

for fitter_results in results:
    mean_errors = []
    for frs in fitter_results:
        errors = [fr.final_error() for fr in frs]
        mean_errors.append(np.mean(errors))
    all_errors.append(mean_errors)
    
all_errors = np.asarray(all_errors)
x_values = [-0.01] + noise_std[1:]
```

```python
plot(x_values, all_errors.T, linewidth=1, marker='o', ms=5)

legend_entries = ['SIC', 'SFC', 
                  'AIC', 'AFC',  
                  'BIC', 'BFC', 'PIC']
legend(legend_entries, bbox_to_anchor=(1.3, 1.025))
xlabel('Initialization')
ylabel('Mean Normalized Point-to-Point Error')
```

# Explore Results

### CED per Algorithm

```python
j = 0

errors = [[fr.initial_error() for fr in results[j][0]]]
for fitter_results in results[j]:
    errors.append([fr.final_error() for fr in fitter_results])
```

```python
from menpofit.visualize import plot_ced

legend_entries = ['gt-projection',
                  'gt',
                  '0.00',
                  '0.02',
                  '0.04',
                  '0.06']

plot_ced(errors[1:], legend_entries=legend_entries[1:])
```

### Statistics per Algorithm

```python
from __future__ import division

print '\t', 'Mean \t', 'STD \t', 'Median \t', 'Convergence \t'

for err, method in zip(errors, legend_entries):
    print method, '\t', 
    print np.round(np.mean(err), decimals=4), '\t', 
    print np.round(np.std(err), decimals=4), '\t', 
    print np.round(np.median(err), decimals=4), '\t',
    
    c = 0
    for e, ini_e in zip(err, errors[0]):
        if e < ini_e:
            c+=1
        
    print np.round(c / len(err), decimals=4)
```
